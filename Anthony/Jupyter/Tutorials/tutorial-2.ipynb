```python
import numpy as np
from scipy import signal

test_array = np.array([1,2,3,4])

def detrend_filter(input_array):
    
    detrended = signal.detrend(input_array)
    return detrended

def detrend(input_array):
    average = np.average(input_array)
    standard_deviation = np.std(input_array)
    subtract = np.subtract(input_array,average)
    detrended = subtract/standard_deviation
    return detrended
```

```python
import matplotlib.pyplot as plt
detrend_array = detrend_filter(test_array)
figure, graph = plt.subplots(2, 1)
graph[0].set_ylabel("RGB-value")
graph[0].set_xlabel("Frame")
graph[0].stem(test_array)
graph[1].set_ylabel("RGB value")
graph[1].set_xlabel("Frame")
graph[1].stem(detrend_array)
plt.grid()
plt.show()
```

```python
def moving_average(values,window):
    weights = np.repeat(1.0,window) / window
    smas = np.convolve(values,weights, 'valid')
    return smas
```

```python
detrend_array = detrend(test_array)
test_ma = moving_average(detrend_array, 2)
test_signal = test_array

figure, graph = plt.subplots(3, 1)
graph[0].set_ylabel("RGB-value")
graph[0].set_xlabel("Frame")
graph[0].stem(test_signal)
graph[1].set_ylabel("RGB value")
graph[1].set_xlabel("Frame")
graph[1].stem(detrend_array)
graph[2].set_ylabel("RGB value")
graph[2].set_xlabel("Frame")
graph[2].stem(test_ma)
plt.grid()
plt.show()
```
